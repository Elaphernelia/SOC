```python
%matplotlib inline

import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## 1. Visualización básica y características generales

```python
#Cargamos la red como edgelist y la visualizamos de manera básica. 
#Cada nodo es una persona y sus nombres vienen dados en terrorists_node_map.txt
G = nx.read_edgelist("terrorists.net", data=False);
```

```python
plt.figure(figsize=(12,9))
nx.draw_networkx(G)
```

```python
#La red superior está visualizada de manera demasiado simple. Vamos a modificarla un poco cambiando el color y tamaño del 
#nodo según su grado. Cuanto más claro y más grande el nodo, mayor es su grado.
plt.figure(figsize=(13,10))

node_color = [G.degree(v) for v in G]
node_size = [200*G.degree(v) for v in G]

nx.draw_networkx(G, node_color=node_color, node_size=node_size, alpha=0.7, edge_color='.4')

```

```python
#Los nodos que parecen que tienen mayor grado son los nodos 0, 6 y 29. Veamos el número exacto de links.
print("Links del nodo 0: ", G.degree('0'))
print("Links del nodo 6: ", G.degree('6'))
print("Links del nodo 29: ", G.degree('29'))
```

```python
#Por último, vamos a obtener el grado medio
dgDict = dict(G.degree())

average_degree = sum(dgDict.values()) / len(dgDict)
print("Grado medio de la red: " , average_degree)
```

## 2. Medidas de centralidad

```python
# Centralidad de grado
degCent = nx.degree_centrality(G)

for x in G.nodes:
  print("Centralidad de grado normalizada del nodo " ,x, ": ", degCent.get(x))
```

```python
# Centralidad de intermediación
btwnCent = nx.betweenness_centrality(G)

for x in G.nodes:
  print("Centralidad de intermediación del nodo " ,x, ": ", btwnCent.get(x))
```

```python
# Centralidad de cercanía
btwnClose = nx.closeness_centrality(G)

for x in G.nodes:
  print("Centralidad de cercanía del nodo " ,x, ": ", btwnClose.get(x))
```

```python
# Centralidad de excentricidad
excentricity = nx.eccentricity(G)

for x in G.nodes:
  print("Excentricidad del nodo " ,x, ": ", excentricity.get(x))
```

## 3. Creación de dataframe

```python
df = pd.read_csv('terrorists_node_map.txt',header=None, sep = " ")
df['first_second_names'] = df[1].str.cat(df[2],sep=" ")
df['first_second_third_names'] = df['first_second_names'].str.cat(df[3],sep=" ")
df['Nombres'] = ""
df['Centralidad de grado'] = ""
df['Centralidad de intermediación'] = ""
df['Centralidad de cercanía'] = ""
df['Excentricidad'] = ""
#Añadimos los nombres de manera que la columna de nombre tenga el nombre completo:
y = range(34)
for x in y:
    if pd.isnull(df.at[x,'first_second_third_names']):
        df.at[x,'Nombres']=(df['first_second_names'].iloc[x])
    else:
        df.at[x,'Nombres']=(df['first_second_third_names'].iloc[x])
        
#Añadimos los valores de las distintas centralidades:      
for dc in G.nodes:
        y = int(dc)
        df.at[y,'Centralidad de grado']=degCent.get(dc)

for bc in G.nodes:
        y = int(bc)
        df.at[y,'Centralidad de intermediación']=btwnCent.get(bc)
        
for cc in G.nodes:
        y = int(cc)
        df.at[y,'Centralidad de cercanía']=btwnClose.get(cc)
        
for e in G.nodes:
        y = int(e)
        df.at[y,'Excentricidad']=excentricity.get(e)
    
df.drop([0, 1, 2, 3, 4, 'first_second_names', 'first_second_third_names'], axis=1)
#Lo que se muestra es la tabla que nos sirve. Por cada nodo se indica el nombre de la persona y sus centralidades
```

## 4. Nodos valiosos

```python
#A continuación vamos a ver los 5 nodos más importantes respecto a las distintas centralidades y compararlos:

df.drop([0, 1, 2, 3, 4, 'first_second_names', 'first_second_third_names'], axis=1).head()
```

```python
#NODOS IMPORTANTES EN CUANTO A CENTRALIDAD DE CERCANÍA

df.sort_values('Centralidad de cercanía', ascending=False).drop([0, 1, 2, 3, 4, 'first_second_names', 'first_second_third_names', 'Centralidad de grado','Centralidad de intermediación','Excentricidad'], axis=1).head()
```
