```python
lista_fecha = []
lista_artistas = []

for x in range(22961):
  lista_cancion = []
  spot_track = "spotify:track:"
  search_uri = df.iloc[x]['spotify_track_id']
  result = sp.track(spot_track + search_uri)
  lista_fecha.append(result['album']['release_date'])
  for i in range(len(result['artists'])):
    lista_cancion.append(result['artists'][i]['name'])
  lista_artistas.append(lista_cancion)

df['release_date'] = lista_fecha
df['artist_list'] = lista_artistas
```
